```python
import json
```

```python
with open("channels_code_full.json") as f:
   chan_list = json.load(f)
f.close()

for i in chan_list:
  
   i['Simple_code'] = i['Code'].replace('.', '')
   i['Latex'] = i['Simple_code'][:-1] + "_{" + i['Simple_code'][-1] + "}"
   i['Prefix'] = i['Simple_code'][:-1]
   i["Underscore"] = i['Simple_code'][-1]

with open('channels_code_proc.json', 'w') as json_file:
  json.dump(chan_list, json_file, indent = 4, sort_keys=True)


len(chan_list), chan_list

print(json.dumps(chan_list, indent = 4, sort_keys=True))

```
