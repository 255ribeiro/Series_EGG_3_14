# Collecting data results

```python
import json
import pandas as pd
from  pathlib import Path
import numpy as np
```

```python
pd.set_option('max_column', None)
pd.set_option('max_row', None)
```

```python
root = Path()
info_path = root / 'info'
data_proc = root / 'data' / 'data_proc'
out_path =root / 'output'
out_path.mkdir(exist_ok=True)
out_tables = out_path / 'exracted_tables'
out_tables.mkdir(exist_ok=True)
```

```python
exp_dict_file = info_path / 'experiment_dict.json'
with open(exp_dict_file) as f:
   exp_dict = json.load(f)
f.close()
exp_dict
```

## Agreegating tables

```python
df = pd.DataFrame()
# for every subject
for k, subj_path in enumerate(data_proc.glob('*/')):
    #print(k, subj_path)
    #For every experiment
    for j , exp_path in enumerate(subj_path.glob('*/')):
            #print(j, exp_path)
            # for every DMCX2 file
            for f , file in enumerate(exp_path.glob('*DMCX2*')):
                info_list = file.stem.split('_')
                #print('\t\t', file,  info_list)
                df_aux = pd.read_csv(file, sep=' ', names=[info_list[3]])
                exp_dir = exp_dict[info_list[1]]['Activity'].split(' ')
                exp_dir = "".join([exp_dir[0][0], exp_dir[1][0]])
                #print(exp_name)
                # print(info_list)
                # print(exp_dict[info_list[1]]['Activity'])
                # print(exp_dict[info_list[1]]['Type'])
                df_aux.columns = pd.MultiIndex.from_product([ [info_list[0]], [info_list[1]], [exp_dict[info_list[1]]['Activity']] , [exp_dict[info_list[1]]['Type']] , [info_list[3]] ], names=["Subject", "Experiment", 'Activity' , 'Type' , 'Channel'])
                df = pd.concat([df,df_aux], axis=1)
df.index.name = 'Time Window'
df.head()
```

```python
df_full = df.drop('S106', axis=1)
df_full.columns = df_full.columns.remove_unused_levels()
print(df_full.shape)
df_full.info()
```

## Saving colected data

```python
file_name = out_tables/'collected_data.xlsx'
df.to_excel(file_name)
file_name = out_tables/'collected_data.pkl'
df.to_pickle(file_name)
print(df.shape)
df.info()

file_name = out_tables/'collected_data_complite.xlsx'
df_full.to_excel(file_name)
file_name = out_tables/'collected_complite.pkl'
df_full.to_pickle(file_name)
print(df_full.shape)
df_full.info()
```

## Stats to be processed

```python
# stats to calc

def std_pop(x):
    return pd.Series.std(x, ddof=0)

stat_func_list = ['mean', 'median', 'std', std_pop ]
stat_func_names = ['mean', 'median', 'std', 'std pop' ]
```

### List of experiments to calculate stats

```python
exp_list= df.columns.levels[1]
exp_list
```

### List of channels to claculate stats

```python
channels = df.columns.levels[4]
channels
```

## Stats of the full table

```python
df_stats = pd.DataFrame()

for experiment in exp_list:

    for i in channels:
        #print( experiment, i)
        df_aux = df.loc[:, (slice(None), [experiment], [i])].agg(stat_func_list, axis=1)
        #print(df_aux.head())
        df_aux.columns = pd.MultiIndex.from_product([[experiment],[i], stat_func_names], names=[ "Experiment",'Channel',   'stats'])
        df_stats = pd.concat([df_stats,df_aux], axis=1)#
df_stats.head()
```
